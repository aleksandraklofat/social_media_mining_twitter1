```python
import tweepy
```

```python
# Eingabe der Passwörter
CONSUMER_KEY = "xyz"
CONSUMER_SECRET = "xyz"
OAUTH_TOKEN = "xyz"
OAUTH_TOKEN_SECRET = "xyz"
```

```python
# Authorisieren beim Twitter via OAuth
auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(OAUTH_TOKEN, OAUTH_TOKEN_SECRET)
```

```python
twitter_api = tweepy.API(auth)
```

```python
# Überprüfen, ob Twitter die Zugangsdaten akzeptiert hat
print(twitter_api)
```

```python
woeid = 23424829 # WOEID für Deutschland
```

```python
# Abrufen von Trends mit entsprechendem WOEID
trends = twitter_api.trends_place(id = woeid)
```

```python
trends
```

```python
# Extrahieren von Trends 
for value in trends:
    for trend in value['trends']:
        print(trend['name'])
```
